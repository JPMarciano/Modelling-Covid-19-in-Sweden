```python
a11=0.25
a21=0.25
t11=100
t21=t11+14
Nprime1 = 9000
N=10230000
time = np.arange(0, 380, 1);
S1=[]
R1=[]
I1=[]
newcases1=[NaN]



for t in time:
    S1.append(N-Nprime1/(1+np.exp(-a11*(t-t11))))
    R1.append(Nprime1/(1+np.exp(-a21*(t-t21))))
    I1.append(N-S1[-1]-R1[-1])
    if t!=0:
        newcases1.append(S1[-2]-S1[-1])
        
        
a12=0.15
a22=0.15
t12=125
t22=t12+14
Nprime2 = 20000
N=10230000
S2=[]
R2=[]
I2=[]
newcases2=[NaN]

for t in time:
    S2.append(N-Nprime2/(1+np.exp(-a12*(t-t12))))
    R2.append(Nprime2/(1+np.exp(-a22*(t-t22))))
    I2.append(N-S2[-1]-R2[-1])
    if t!=0:
        newcases2.append(S2[-2]-S2[-1])
        
a13=0.10
a23=0.10
t13=175
t23=t13+14
Nprime3 = 50000
N=10230000
S3=[]
R3=[]
I3=[]
newcases3=[NaN]

for t in time:
    S3.append(N-Nprime3/(1+np.exp(-a13*(t-t13))))
    R3.append(Nprime3/(1+np.exp(-a23*(t-t23))))
    I3.append(N-S3[-1]-R3[-1])
    if t!=0:
        newcases3.append(S3[-2]-S3[-1])
        
        
a14=0.08
a24=0.08
t14=325
t24=t14+14
Nprime4 = 300000
N=10230000
S4=[]
R4=[]
I4=[]
newcases4=[NaN]

for t in time:
    S4.append(N-Nprime4/(1+np.exp(-a14*(t-t14))))
    R4.append(Nprime4/(1+np.exp(-a24*(t-t24))))
    I4.append(N-S4[-1]-R4[-1])
    if t!=0:
        newcases4.append(S4[-2]-S4[-1])


newcases=[newcases1[i]+newcases2[i]+newcases3[i]+newcases4[i] for i in range(380)]
        
plt.plot(time[0:333], cleandf[['new_cases_smoothed']])
plt.plot(time, newcases1)
plt.plot(time, newcases2)
plt.plot(time, newcases3)
plt.plot(time, newcases4)


plt.title('newcases')

plt.xlabel('Time')

plt.grid(True, which='both')

plt.axhline(y=0, color='k')

plt.show()
```
